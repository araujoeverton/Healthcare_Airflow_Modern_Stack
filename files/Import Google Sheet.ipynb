# Importing a Google Sheet

## Authorizing access to your Google Drive

```
from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
gc = gspread.authorize(GoogleCredentials.get_application_default())
```

## Importing the Sheet

```
wb = gc.open('correct').sheet1  # Importing the Sheet
```

```
data = wb.get_all_values()  # Reading all the values
```

## Saving the data as a `pandas` `DataFrame` object


```
from pandas import DataFrame
```

```
df = DataFrame(data)  # Saving the read values as a DataFrame object
```

```
df.head()  # Viewing the first 5 rows
```

```
df.columns = df.iloc[0]  # Setting row 1 as the column headers (variable names)
```

```
df.head()
```

```
df.drop(0, inplace=True)  # Permanently deleting row 1
```
